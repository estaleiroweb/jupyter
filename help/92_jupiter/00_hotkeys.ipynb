## Atalhos úteis

- `Esc + A`: Inserir nova célula acima.
- `Esc + B`: Inserir nova célula abaixo.
- `Esc + D + D`: Deletar célula.
- `Esc + M`: Muda célula para Markdown.
- `Esc + Y`: Muda célula para código.
- `Shift + Enter`: Executa a célula.
- `Ctrl + Shift + -`: Divide célula no cursor.
