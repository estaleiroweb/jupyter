# While / For Loops

```python
i = 0

while i < 6:
    print(i)
    i += 1
    if i == 9:
        continue
    if i == 3:
        break
```

```python
fruits = ["apple", "banana", "cherry"]
for x in fruits:
	print(x)
	if x == "banana":
		break
	if x == "apple":
		continue
```

```python
for x in "banana":
	print(x)
```

```python
for x in range(6):
	print(x)
```

```python
for x in range(2,6):
	print(x)
```

```python
for x in range(2, 30, 3):
	print(x)
```

```python
for x in range(6): #Print all numbers from 0 to 5, and print a message when the loop has ended
	print(x)
else:
	print("Finally finished!")
```
